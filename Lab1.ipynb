```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def lorenz_attractor(sigma, rho, beta, initial_state, num_steps):
    dt = 0.01
    x, y, z = initial_state
    x_vals, y_vals, z_vals = [x], [y], [z]

    for _ in range(num_steps):
        dx = sigma * (y - x)
        dy = x * (rho - z) - y
        dz = x * y - beta * z

        x += dt * dx
        y += dt * dy
        z += dt * dz

        x_vals.append(x)
        y_vals.append(y)
        z_vals.append(z)

    return x_vals, y_vals, z_vals

# Parameters for the Lorenz attractor
sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

# Initial state and number of time steps
initial_state = [0.1, 0.0, 0.0]
num_steps = 10000

# Generate Lorenz attractor data
x_vals, y_vals, z_vals = lorenz_attractor(sigma, rho, beta, initial_state, num_steps)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the Lorenz attractor
ax.plot(x_vals, y_vals, z_vals, lw=0.5)

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Lorenz Attractor')

plt.show()
```

```python
# Beam.py: solves Navier−−Stokes equation for flow around beam
import matplotlib.pylab as p
from mpl_toolkits.mplot3d import Axes3D
from numpy import *

print("Working, wait for the figure after 100 iterations")

Nxmax = 70
Nymax = 20
IL = 10
H = 8
T = 8
h = 1.0

u = zeros((Nxmax+1, Nymax+1), float)  # Stream
w = zeros((Nxmax+1, Nymax+1), float)  # Vorticity

V0 = 1.0
omega = 0.1
nu = 1.0
iter = 0
R = V0 * h / nu

def borders():
    # Initialize boundary conditions
    for i in range(0, Nxmax+1):
        for j in range(0, Nymax+1):
            w[i, j] = 0.
            u[i, j] = j * V0

    # Fluid surface
    for i in range(0, Nxmax+1):
        u[i, Nymax] = u[i, Nymax-1] + V0 * h
        w[i, Nymax-1] = 0.

    # Inlet
    for j in range(0, Nymax+1):
        u[1, j] = u[0, j]
        w[0, j] = 0.

    # Centerline
    for i in range(0, Nxmax+1):
        if IL <= i <= IL+T:
            u[i, 0] = 0.
            w[i, 0] = 0.

    # Outlet
    for j in range(1, Nymax):
        w[Nxmax, j] = w[Nxmax-1, j]
        u[Nxmax, j] = u[Nxmax-1, j]

def beam():
    # Boundary conditions for the beam
    for j in range(0, H+1):
        # Front
        w[IL, j] = -2 * u[IL-1, j] / (h * h)
        # Back
        w[IL+T, j] = -2 * u[IL+T+1, j] / (h * h)

    for i in range(IL, IL+T+1):
        w[i, H-1] = -2 * u[i, H] / (h * h)

    for i in range(IL, IL+T+1):
        for j in range(0, H+1):
            # Front and Back
            u[IL, j] = 0.
            u[IL+T, j] = 0.
            # Top
            u[i, H] = 0.

def relax():
    # Relaxation of the stream and vorticity
    beam()  # Reset conditions
    for i in range(1, Nxmax):
        for j in range(1, Nymax):
            r1 = omega * ((u[i+1, j] + u[i-1, j] + u[i, j+1] + u[i, j-1] + h * h * w[i, j]) / 4 - u[i, j])
            u[i, j] += r1

    for i in range(1, Nxmax):
        for j in range(1, Nymax):
            a1 = w[i+1, j] + w[i-1, j] + w[i, j+1] + w[i, j-1]
            a2 = (u[i, j+1] - u[i, j-1]) * (w[i+1, j] - w[i-1, j])
            a3 = (u[i+1, j] - u[i-1, j]) * (w[i, j+1] - w[i, j-1])
            r2 = omega * ((a1 - (R/4.) * (a2 - a3)) / 4. - w[i, j])
            w[i, j] += r2

    borders()

while iter <= 100:
    iter += 1
    if iter % 10 == 0:
        print(iter)
    relax()

# Normalize the stream function
for i in range(0, Nxmax+1):
    for j in range(0, Nymax+1):
        u[i, j] = u[i, j] / (V0 * h)

# Plot the stream function
x = range(0, Nxmax-1)
y = range(0, Nymax-1)
X, Y = p.meshgrid(x, y)

def functz(u):
    # Stream function
    z = u[X, Y]
    return z

Z = functz(u)
fig = p.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, Z, color='r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Stream Function')
p.show()
```
